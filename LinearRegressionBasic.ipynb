```python
weight=[2,4,5,3,6,5,7]
price=[35, 60, 20, 50, 50, 55, 60]
```

```python
df=pd.DataFrame({'Weight': weight, 'Price': price}, columns=['Weight', 'Price'])
df
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(df[['Weight']],df['Price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
result=(reg.coef_*6)+reg.intercept_
result
```

```python
df['Predicted Price']=reg.predict(df[['Weight']])
df
```

```python
df['Residual Error']=df['Price']-df['Predicted Price']
df
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(df['Price'], df['Predicted Price'])
mse
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(df['Price'], df['Predicted Price'])
mae
```
