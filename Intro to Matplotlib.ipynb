## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt 

#1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. Setup plot
fig, ax=plt.subplots(figsize=(10,10)) # width, height

#3. Plot data
ax.plot(x,y)

#4. Cutomize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5. Save & Show 
fig.savefig("./sample_plot.png")
```

```python
### Which one should I use? pyplot vs matplotlib OO Method?

* When plotting something quickly --> pyplot method
* When plotting something more advanced --> OO Method
```

```python
over_50=heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO Method

fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100])
```

```python
## OO Method from scratch
fig, ax=plt.subplots(figsize=(10,6))

# Plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age",ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"]);

# Customize the plot ax0
ax0.set(title="Heart Disease and Cholesterol Levels",ylabel="Cholesterol");

# Add a legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');


# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"]);

# Customize the plot ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age",ylabel="Thalach");

# Add a legend ax1
ax1.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle='--');

# Add a title to the fig
fig.suptitle("Heart Disease Analysis", fontsize='16', fontweight='bold');
```

```python
# Use plot style
plt.style.use('seaborn-v0_8-whitegrid')

## OO Method from scratch
fig, ax=plt.subplots(figsize=(10,6))

# Plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"],
                  cmap="winter") # changes color

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age",ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# Customizing



# Subplot of chol, age, thalach

fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True);

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"],
                   cmap="winter");

# Customize the plot ax0
ax0.set(title="Heart Disease and Cholesterol Levels",ylabel="Cholesterol");

# Change x-axis limit
ax0.set_xlim([50,80])
# Add a legend ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');


# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"],
                   cmap="winter");

# Customize the plot ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age",ylabel="Thalach");

# Change x-axis & y-axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])


# Add a legend ax1
ax1.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle='--');

# Add a title to the fig
fig.suptitle("Heart Disease Analysis", fontsize='16', fontweight='bold');
```
