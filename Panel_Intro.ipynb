```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
```

```python
df=pd.read_csv("./merge.csv")
df
```

```python
idf=df.interactive()
```

```python
year_slider=pn.widgets.IntSlider(name='Year Slider',start=2016,end=2022,step=1,value=7)
year_slider
```

```python
df.tail()
```

```python
df["Year"]='0'
```

```python
#df["Year"]=df[df['Date (LT)'].str.contains('2016')]
```

```python
df.loc[df["Date (LT)"].str.contains('2016'), "Year"] = "2016"
```

```python
df.loc[df["Date (LT)"].str.contains('2017'), "Year"] = "2017"
df.loc[df["Date (LT)"].str.contains('2018'), "Year"] = "2018"
df.loc[df["Date (LT)"].str.contains('2019'), "Year"] = "2019"
df.loc[df["Date (LT)"].str.contains('2020'), "Year"] = "2020"
df.loc[df["Date (LT)"].str.contains('2021'), "Year"] = "2021"
df.loc[df["Date (LT)"].str.contains('2022'), "Year"] = "2022"
```

```python
df
```

```python
df["Year"] = df["Year"].astype(int)
```

```python
df["Month"]='0'
```

```python
df.loc[df["Date (LT)"].str.contains('/01/'), "Month"] = "January"
df.loc[df["Date (LT)"].str.contains('/02/'), "Month"] = "February"
df.loc[df["Date (LT)"].str.contains('/03/'), "Month"] = "March"
df.loc[df["Date (LT)"].str.contains('/04/'), "Month"] = "April"
df.loc[df["Date (LT)"].str.contains('/05/'), "Month"] = "May"
df.loc[df["Date (LT)"].str.contains('/06/'), "Month"] = "June"
df.loc[df["Date (LT)"].str.contains('/07/'), "Month"] = "July"
df.loc[df["Date (LT)"].str.contains('/08/'), "Month"] = "August"
df.loc[df["Date (LT)"].str.contains('/09/'), "Month"] = "September"
df.loc[df["Date (LT)"].str.contains('/10/'), "Month"] = "October"
df.loc[df["Date (LT)"].str.contains('/11/'), "Month"] = "November"
df.loc[df["Date (LT)"].str.contains('/12/'), "Month"] = "December"
```

```python
radio_group = pn.widgets.RadioButtonGroup(
    name='Y-Axis', 
    options=['AQI'], 
    button_type='success')

radio_group
```

```python
Month=["January","February","March","April","May","June","July","August","September","October","November","December"]

AQI_Pipeline=(
    idf[
        (idf.Year == year_slider) &
        (idf.Month.isin(Month))
    ]
    .groupby(['Month', 'Year','Date (LT)'])[radio_group].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
AQI_Pipeline
```

```python
AQI_plot = AQI_Pipeline.hvplot(x = 'Month', by='Month', y=radio_group,line_width=2, title="Yearly AQI Value over 12 months",width=1000)
AQI_plot
```

```python
AQI_vs_Date_scatterplot_pipeline = (
    idf[
        (idf.Year == year_slider) &
        (idf.Month.isin(Month))
    ]
    .groupby(['Month', 'Year', 'Date (LT)'])[radio_group].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
AQI_vs_Date_scatterplot_pipeline
```

```python
AQI_vs_Date_scatterplot = AQI_vs_Date_scatterplot_pipeline.hvplot(x='Month', 
                                                                y='AQI', 
                                                                by='Year', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=True, 
                                                                height=500, 
                                                                width=1000)
AQI_vs_Date_scatterplot
```

```python
AQI_Diff = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['AQI', 'AQI Category'], 
    button_type='success'
)

Month=["January","February","March","April","May","June","July","August","September","October","November","December"]

AQI_bar_pipeline = (
    idf[
        (idf.Year == year_slider) &
        (idf.Month.isin(Month))
    ]
    .groupby(['Month', 'Year', 'Date (LT)'])[AQI_Diff].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='Year')  
    .reset_index(drop=True)
)
```

```python
AQI_bar_plot = AQI_bar_pipeline.hvplot(kind='bar', 
                                        x='Month', 
                                        y=AQI_Diff,
                                      by='Year', 
                                        title='AQI Status of Different Months',
                                      width=900)
AQI_bar_plot
```

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='Dhaka Air Quality Index over last 7 Years dashboard', 
    sidebar=[pn.pane.Markdown("# AQI Value and Dhaka Air Pollution"), 
             pn.pane.Markdown("#### Dhaka is subject to some fairly bad pollution levels annually, with only brief periods of respite, that despite being lower still hold some relatively high numbers of pollution readings, making its air harmful to breathe year-round."), 
             pn.pane.PNG('air.jpg', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"),   
             year_slider],
    main=[pn.Row(pn.Column(radio_group, 
                           AQI_plot.panel(width=700), margin=(0,25))), 
          pn.Row(pn.Column(AQI_vs_Date_scatterplot.panel(width=600), margin=(0,25)), 
                 pn.Column(AQI_bar_plot.panel(width=600)))],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
# template.show()
template.servable();
```
